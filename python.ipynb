```python
#16. Implement a function to check if a given list is sorted in non-decreasing order.
def is_sorted(nums):
    # Check if the list has 0 or 1 element
    if len(nums) <= 1:
        return True

    # Iterate through the list from the second element onwards
    for i in range(1, len(nums)):
        # Check if the current element is smaller than the previous element
        if nums[i] < nums[i - 1]:
            return False

    # All elements are in non-decreasing order
    return True

numbers = [1, 2, 3, 4, 5]
if is_sorted(numbers):
    print("The list is sorted.")
else:
    print("The list is not sorted.")

```

```python
#28. Implement a function to calculate the square root of a given number.
def square_root(num, precision=0.0001):
    if num < 0:
        raise ValueError("Square root is not defined for negative numbers.")

    if num == 0:
        return 0

    x = num
    while True:
        root_candidate = 0.5 * (x + num / x)
        if abs(x - root_candidate) < precision:
            return root_candidate
        x = root_candidate

# Test the function
number = float(input("Enter a number: "))
result = square_root(number)
print("Square root of", number, "is:", result)
```

```python
#30. Implement a function to find the minimum element in a rotated sorted list.
def find_minimum(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        # Check if mid element is greater than the rightmost element
        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    # The minimum element will be at the left index
    return nums[left]

# Test the function
nums = [6,5,4,3,2,1]
result = find_minimum(nums)
print("Minimum element in the list:", result)
```

```python
#34. Implement a function to find the longest common prefix among a list of strings.
def longest_common_prefix(strs):
    if not strs:
        return ""

    prefix = strs[0]
    for string in strs[1:]:
        while string.find(prefix) != 0:
            prefix = prefix[:-1]
            if not prefix:
                return ""
    return prefix

# Test the function
str_list = input("Enter a list of strings separated by spaces: ").split()
result = longest_common_prefix(str_list)
print("Longest common prefix:", result)
```

```python
#35. Write a program to check if a given number is a perfect square.
def is_perfect_square(num):
    if num < 0:
        return False

    # Using binary search to find the square root of the number
    left, right = 0, num
    while left <= right:
        mid = left + (right - left) // 2
        square = mid * mid

        if square == num:
            return True
        elif square < num:
            left = mid + 1
        else:
            right = mid - 1

    return False

# Test the function
number = int(input("Enter a number: "))
if is_perfect_square(number):
    print(number, "is a perfect square.")
else:
    print(number, "is not a perfect square.")

    
```

```python
#41. Write a Python program to find the smallest missing positive integer in a list.
def find_smallest_missing_positive(nums):
    n = len(nums)

    # Step 1: Move all positive integers to the left side
    i = 0
    while i < n:
        if nums[i] > 0 and nums[i] <= n and nums[nums[i] - 1] != nums[i]:
            nums[nums[i] - 1], nums[i] = nums[i], nums[nums[i] - 1]
        else:
            i += 1

    # Step 2: Find the first index where nums[i] != i + 1
    for i in range(n):
        if nums[i] != i + 1:
            return i + 1

    # Step 3: If all indices are correct, the missing positive is n + 1
    return n + 1

# Test the program
input_list = [int(x) for x in input("Enter a list of integers separated by spaces: ").split()]
result = find_smallest_missing_positive(input_list)
print("Smallest missing positive integer:", result)
```

```python
#42. Implement a function to find the longest palindrome substring in a given string.
def longest_palindrome_substring(s):
    if not s:
        return ""

    longest_palindrome = ""
    n = len(s)

    # Function to expand around a center for odd length palindrome
    def expand_around_center(left, right):
        while left >= 0 and right < n and s[left] == s[right]:
            left -= 1
            right += 1
        return s[left + 1:right]

    for i in range(n):
        # Check for odd length palindrome
        palindrome_odd = expand_around_center(i, i)
        if len(palindrome_odd) > len(longest_palindrome):
            longest_palindrome = palindrome_odd

        # Check for even length palindrome
        palindrome_even = expand_around_center(i, i + 1)
        if len(palindrome_even) > len(longest_palindrome):
            longest_palindrome = palindrome_even

    return longest_palindrome

# Test the function
input_string = input("Enter a string: ")
result = longest_palindrome_substring(input_string)
print("Longest palindrome substring:", result)
```

```python
#44. Implement a function to check if a given number is a perfect number.
def is_perfect_number(number):
    if number <= 0:
        return False

    divisor_sum = 0
    for i in range(1, number):
        if number % i == 0:
            divisor_sum += i

    return divisor_sum == number

# Test the function
number = int(input("Enter a number: "))
if is_perfect_number(number):
    print("The number is a perfect number.")
else:
    print("The number is not a perfect number.")
```

```python
#46. Implement a function to find the first missing positive
def first_missing_positive(nums):
    n = len(nums)

    # Step 1: Separate positive and non-positive integers
    pos_nums = []
    for num in nums:
        if num > 0:
            pos_nums.append(num)

    # Step 2: Mark positive integers in the original list
    for i in range(len(pos_nums)):
        num = abs(pos_nums[i])
        if num <= n:
            nums[num-1] = -abs(nums[num-1])

    # Step 3: Find the first positive integer in the original list
    for i in range(n):
        if nums[i] > 0:
            return i + 1

    # Step 4: If all positive integers are present, the missing positive is n + 1
    return n + 1

# Test the function
input_list = [int(x) for x in input("Enter a list of integers separated by spaces: ").split()]
result = first_missing_positive(input_list)
print("First missing positive integer:", result)
```
